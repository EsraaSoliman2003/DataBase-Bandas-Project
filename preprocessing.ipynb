## Outliers

```python
df = pd.read_csv("data/age_prediction.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df.boxplot()
plt.show()
```

```python
Q1 = df["Age"].quantile(.25)
Q3 = df["Age"].quantile(.75)
IQR = Q3-Q1
upper_limit = Q3 + 1.5*IQR
lower_limit = Q1 - 1.5*IQR
```

```python
df.loc[df["Age"]>upper_limit,"Age"] = df["Age"].mean()
```

```python
df.loc[df["Age"]<lower_limit,"Age"] = df["Age"].mean()
```

```python
df.columns
```

```python
num_cols = df.select_dtypes(include=np.number).columns
num_cols
```

```python
for col in num_cols:
    Q1 = df[col].quantile(.25)
    Q3 = df[col].quantile(.75)
    IQR = Q3-Q1
    upper_limit = Q3 + 1.5*IQR
    lower_limit = Q1 - 1.5*IQR
    df.loc[df[col]>upper_limit,col] = df[col].mean()
    df.loc[df[col]<lower_limit,col] = df[col].mean()
```

```python
df.boxplot()
plt.show()
```
